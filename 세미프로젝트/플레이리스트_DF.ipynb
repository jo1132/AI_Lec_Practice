#함수 미리 정의

```python
def Please_Insert(data_frame, col):
  item_list = data_frame[col].values.tolist()

  items = []
  for itm in item_list:
    items += itm.split(' ')

  #중복제거
  set_items = set(items)
  data_frame[col].iloc[0] = (' ').join(list(set_items))
  return data_frame
```

```python
def to_str(num):
  if not (type(num) == str):
    num = round(num)
    num = str (num)
  return num
```

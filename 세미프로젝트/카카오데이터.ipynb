##. 1. 불러오기

```python
song_df = pd.read_json('/content/drive/MyDrive/Colab Notebooks/세미프로젝트/data/song_meta.json')
pyl_df = pd.read_json('/content/drive/MyDrive/Colab Notebooks/세미프로젝트/data/train.json')
```

## 2. 인덱스 만들기
> loc를 사용할 수 있도록 만들어준다.

```python
song_df['index'] = song_df['id']
song_df = song_df.set_index('index')
song_df
```

## 각 노래에 태그 넣기

```python
song_df['태그'] = ''
song_df
```

```python
def addsharp(item):
  if (item[0] != '#'):
    item = '#'+item
  return item
```

```python
i = 0
for pyl in pyl_df.index:
  song_list = pyl_df.loc[pyl]['songs']
  for song_item in song_list:
    tag_list = pyl_df.loc[pyl]['tags']
    tag_list += song_df.loc[song_item]['태그']
    tag_list = list(map(addsharp, tag_list))

    tag_set = set(tag_list)
    if '' in tag_set:
      tag_set.remove('')
  
    tag_list = list(tag_set)

    song_df['태그'].loc[song_item]  = tag_list
  i += 1
  print(song_df.loc[song_item]['태그'], i, '/', len(pyl_df.index), 'done')
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/세미프로젝트/data/Base_Song_meta.csv')
df
```

```python
df['태그'] = df['태그'].astype(str)
df['index'] = df['index'].astype(str)
df = df.set_index('index')
```

```python
df['태그리스트'] = df['태그'].apply((lambda x: x.split(' ')))
df
```

```python
tag_list = []
for i in range(len(df['태그리스트'].values)):
  tag_list += df['태그리스트'].iloc[i]
tag_list
```

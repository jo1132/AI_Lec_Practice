# 코사인 유사도 실습

```python
docs = [
        '여름엔 과일이 맛있다', 
        '여름엔 수박이 맛있다',
        '여름엔 바다지',
        '여름엔 과일이 맛없다']
```

```python
import pandas as pd
import numpy as np
```

##1. 데이터 변환

```python
from sklearn.feature_extraction.text import CountVectorizer

vect = CountVectorizer()
count_vect = vect.fit_transform(docs)
count_vect.toarray()
```

##2. 수식 구현

###2.1 분모중 왼쪽 수식 Ai구현

```python
count_vect_arr = count_vect.toarray()
count_vect_arr
```

```python
a = count_vect_arr[0]
val_square = np.square(a)
val_square
```

> a가 가지고 있는 데이터 각각에 제곱을 처리했다.

```python
val_square_sum = sum(val_square)
val_square_sum
```

> 각 데이터의 합을 구했다.

```python
ai = np.sqrt(val_square_sum)
ai
```

> 합에 대한 루트값을 구했다

##2.2 분자 구하기 

```python
a = count_vect_arr[0]
b = count_vect_arr[1]
numerator = np.dot(a, b)
numerator
```

> 첫 번째 행을 a에 두번 째 행을 b에 저장하고, 각 행의 데이터를 서로 곱한 후 합을 구한 것이다.

### 2.3 Quiz 
>문서1과 문서2의 유사도를 구하시오

```python
def getRootSqrt(arr):
  import numpy as np
  squre = np.square(arr)
  squre_sum = np.sum(squre)
  return np.sqrt(squre_sum)
```

```python
def getSimilarity(a, b):
  import numpy as np
  ai = getRootSqrt(a)
  bi = getRootSqrt(b)
  a_dot_b = np.dot(a, b)
  return a_dot_b / (ai*bi)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vect = CountVectorizer()
count_vect = vect.fit_transform(docs)
array = count_vect.toarray()
array
```

> 문서 1(a)과 문서2(b)의 유사도

```python
a = array[0]
b = array[1]
ai = getRootSqrt(a)
bi = getRootSqrt(b)
a_dot_b = np.dot(a, b)
print("ai:",ai)
print("bi:",bi)
print("a_dot_b:", a_dot_b)
similarity = a_dot_b / (ai*bi)
print("유사도:",similarity)
```

```python
for i in range(len(array)):
  print("문서1과 문서{}의 유사도 = {}".format(i+1, round(getSimilarity(array[0], array[i]), 5)))
```

```python
def getRootSqrt(arr):
  import numpy as np
  return np.sqrt(np.sum(np.square(arr)))

def getSimilarity(a, b):
  import numpy as np
  return np.dot(a, b) / (getRootSqrt(a)*getRootSqrt(b))

def getArray(dataset):
  from sklearn.feature_extraction.text import CountVectorizer
  return CountVectorizer().fit_transform(dataset).toarray()
```

```python
docs2 = [
        '여름엔 과일이 맛있다 여름엔 과일이 맛있다', 
        '여름엔 수박이 맛있다',
        '여름엔 바다지',
        '여름엔 과일이 맛없다'
        ]
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
```

```python
array = getArray(docs2)
#array
for i in range(len(array)):
  for j in range(len(array)):
    print("문서{}과 문서{}의 유사도 = {}".format(i+1, j+1, getSimilarity(array[i], array[j])))
  print("_______________________________________________________")
```

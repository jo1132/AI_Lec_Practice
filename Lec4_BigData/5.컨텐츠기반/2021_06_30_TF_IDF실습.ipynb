#1. 이해하기
> 단순히 빈도수만을 구하는 countvectorizer의 문제점을 보완하여 빈도수와 문서에서 나타나는 빈도수의 역치를 이용한 알고리즘 이다.

```python
docs = [
        '여름엔 과일이 맛있다', 
        '여름엔 수박이 맛있다',
        '여름엔 바다지',
        '여름엔 과일이 맛없다']
```

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vect = TfidfVectorizer()
vect_fit = vect.fit(docs)
vect_transform = vect.transform(docs)
```

```python
df = pd.DataFrame(vect_transform.toarray(), columns=sorted(vect.vocabulary_))
df.index=['문서1', '문서2', '문서3', '문서4']
df
```

> 실행결과 countVectorizer를 이용하면 수치가 계속 증가되어 커지지만 tfidfvectorizer를 이용하면 log처리로 수치를 낮출 수 있으며 단순히 단어의 빈도만 처리하는 것보다 나은 결과를 얻을 수 있다.

```python
from sklearn.metrics.pairwise import cosine_similarity
cosinen_matrix = cosine_similarity(df, df)
cosinen_matrix
```

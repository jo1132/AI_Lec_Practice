##1.5 이상치
> 기존의 데이터와 많은 차이를 보이는 데이터

> IQR(Interquartile Range)의 1.5배 또는 3배의 차이가 나면 이상치라고 표현한다.
* IQR = (3분위수 - 1분위수)

```python
dataset = pd.Series([16, 21, 22, 23, 24, 25, 30])

#percentile(data, 100분위수)
print("0%", np.percentile(dataset, 0))
print("100% : ", np.percentile(dataset, 100))
```

```python
min = np.percentile(dataset, 0)
q1 = np.percentile(dataset, 25)
q2 = np.percentile(dataset, 50)
q3 = np.percentile(dataset, 75)
max = np.percentile(dataset, 100)
```

```python
print("""
min = %.1f
q1 = %.1f
q2 = %.1f
q3 = %.1f
max = %.1f
""" %(min, q1, q2, q3, max))
```

```python
IQR = q3-q1

#최저 한계치
lowfence = q1 - (IQR*1.5)
#최대 한계치
maxfence = q3 + (IQR*3)

print("IQR : ", IQR)

print(dataset)
#최소한계치 출력
print("\n lowfence : ", lowfence)
print(dataset[dataset < lowfence])

#최대 한계치 출력
print("\n maxfence : ", maxfence)
print(dataset[dataset>maxfence])

lowerbar = dataset[dataset>lowfence].min()
maxbar = dataset[dataset<maxfence].max()
```

```python
plt.boxplot(dataset)
plt.text(0.90, lowfence, lowfence)
plt.text(0.8, lowerbar, lowerbar)
plt.text(1.02, min, min)
plt.text(1.09, q1, q1)
plt.text(1.09, q2, q2)
plt.text(1.09, q3, q3)
plt.text(1, maxbar, maxbar)
plt.text(0.9, maxfence, maxfence)
plt.text(1, max, max)

plt.show()
```

###1.5.1 이상치 제거
> 최저한계치보다 작은 값과 최고 한계보다 큰 값을 제거

```python
#dataset
dataset_remove = dataset[dataset>lowfence]
dataset_remove = dataset_remove[dataset_remove<maxfence]
dataset_remove
```

0. 이상치 확인

2. 최대, 최소 한계치 구하기

```python
dataset = pd.Series([16, 21, 22, 23, 24, 25, 3])
dataset.describe()
```

```python
plt.boxplot(dataset)
plt.show
```

```python
q1 = dataset.describe().loc['25%']
q3 = dataset.describe().loc['75%']

#대표값 구하기
median = dataset.describe().loc['50%']

IQR = q3 - q1
low = q1 - IQR*1.5
upper = q3 + IQR*3

print(low, upper, median)

dataset_remove = dataset
dataset_remove[(dataset < low) | (dataset > upper)] = median

dataset_remove
```

```python
plt.boxplot(dataset_remove)
plt.show()
```

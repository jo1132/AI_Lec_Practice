###2.1.2 결측치 확인

> 결측치는 NaN으로 나타나며 isnull()함수를 이용하면 결측치에 대해 True라는 값을 반환하게 된다.

```python
df.isna()
```

```python
df.notnull()
```

> notnull()함수는 결측치가 발생한 경우 False를 반환하며 유효 데이터만 사용하여 분석할 경우 활용된다.

```python
df.isnull().sum()
```

> 결측치에 대한 결과를 확인하기 위해 sum()함수를 이용한다. 결과는 각 feature(속성)별 결과를 보여주기 때문에 한눈에 결측치를 확인할 수 있다.

```python
df.notnull().sum(0)
```

>sum()함수에 전달되는 숫자값은 axis 값으로 행과 열을 뜻하며 0이면 열을,
1이면 행단위로 집계하는 함수이다.

```python
df.notnull().sum(1)
```

```python
series_notnull = df.notnull().sum(1)
print(type(series_notnull))
series_notnull[series_notnull<10]
```

> 위와같이 유효데이터가 적은 값들이 발생되면 행을 제거하여 사용할 수도 있다.

###2.2 결측치 제거
> 결측치 제거는 결측치가 포함되어 있는 행. 열 또는 전체를 제거할 수 있으며 다음과 같이 사용 가능 하다.
* 행 삭제 : df.dropna(axis = 0)
* 열 삭제 : df.dropna(axis = 1)
* 전체삭제: df.dropna()

```python
df
```

```python
df.dropna(axis=0)
```

> 두 결과를 보면 결측치가 발생한 2, 3 행이 삭제되어 나타난 것을 볼 수 잇다.

```python
df
```

> 위의 실행 결과 삭제된 데이터가 반영되지 않았다. 이는 df.dropna(axis=0) 함수를 처리하더라도 결과를 저장하지 않았기 때문에 df에 결과가 반영되지 않은 것이다.

```python
df.dropna(axis=1)
```

> axis가 1인경우 feature 정보에 결측치가 발생되면 제거하게 된다.

```python
print(df['이용거리'][:5])
df['이용거리'].dropna()[:5]
```

> df라는 DataFrame에서 '이용거리'라는 feature만 추출하는데 5개의 데이터만 추출한 결과이고 두 번째 코드는 그 중 결측치가 있으면 제거한 후 5개의 데이터만 추출한 결과 이다.

> 별도의 feature만 처리할 수도 있다. 이를 이용하면 특정 feature의 데이터만 처리하고 나머지 feature에 대해서는 결측치값을 유지할 수도 잇다.

```python
print(df[['대여거치대', '반납거치대', '이용거리']][:5])
```

```python
df[['대여거치대', '반납거치대', '이용거리']].dropna(axis=0)[:5]
```

> 여러 feature들을 이용하여 결측치를 처리할 수도 있다.

##2.3 결측치 대체
> 데이터가 충분하지 않은 경우 하나의 데이터도 소중하다. 이를 유지하기 위해 결측치로 나타나는 데이터를 0, 또는 대표값으로 치환 할 수 있다.

* df.fillna(0) : 결측치를 0으로 대체 (숫자형)
* df.fillna('test') : 결측치를 특정 문자열로 대체 (문자열)
* df.fillna(df.mean()) : 결측치를 평균으로 대체 


```python
df.fillna(0)
```

```python
#df['이용거리'].fillna(0)
df.이용거리.fillna(0)
```

>.(점)은 앞 객체가 가지고 있는 속성을 의미하며 위와같이 다양한 방법으로 표현가능하다.

```python
df.fillna('missing')
```

> 원하는 문자열로 대체가 가능하다.

```python
df_backup = df
df_backup.mean()
```

>DataFrame 에서 평균을 구하게 되면 각 feature별 평균을 구하게 된다.

```python
df_backup.mean()['이용거리']
df_backup['이용거리'].fillna(df_backup.mean()['이용거리'])
```

> 첫 번째 줄에서는 DataFrame의 각 feature별 평균 값 중 '이용거리'만 추출한 것이며 두번째 줄에서는 DataFrame 값 중 '이용거리'feature에 결측치를 평균으로 대체한 것이다.
실행결과 2번 행이 평균값으로 변경된 것을 볼 수 있다.

```python
df_backup.fillna(df_backup.mean()['이용거리'])
df_backup.fillna(df_backup.mean()['이용거리'])['이용거리']
```

> 이전 코드와 같은 결과를 갖게 되지만 현재 코드는 모든 결측치에 영향을 미칠 뿐더러 하드웨어적인 자원의 낭비도 이루어 지기 때문에 비효율적인 방법이긴 하나 여러 방향성을 알아 보는 것으로는 좋다고 볼 수도 잇다.

```python
df_backup['이용거리'] = df_backup['이용거리'].fillna(df_backup.mean()['이용거리'])
df_backup['이용거리'] 
```

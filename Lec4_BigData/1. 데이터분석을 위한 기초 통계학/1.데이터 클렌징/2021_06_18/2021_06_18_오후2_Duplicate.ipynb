#3. 데이터 중복 확인
> DataFrame에서는 중복 확인을 위해 duplicate라는 함수를 지원한다. 중복된 데이터가 나타나면 True를 리턴한다.

```python
df.duplicated(['이용거리'])
```

```python
print(df['이용거리'][:5])
print(df.duplicated(['이용거리'])[:5])
```

> 0번이 False이고 1번이 True이다. 이런경우 1번이 중복된 데이터를 의미한다.

```python
print(df['이용거리'][:5])
df.duplicated(['이용거리', '나이'])
```

> feature가 두 개 이상 사용되면 and연산으로 두 조선이 모두 만족되는 경우만 중복으로 처리한다.

```python
df.duplicated(['이용거리'], keep='last')
```

>keep 옵션은 기본으로 first인데, 첫번째 원소를 빼고 뒤에 나오는 중복 원소가 True

>keep =last 일때, 마지막 원소를 빼고 이전 나왔던 중복 원소들이 True

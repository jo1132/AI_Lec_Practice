#2. 데이터 확인
> 이용시간과 이용거리가 상관관계가 있다는 가정하에 이용시간이 결측치이면 발생된 결측치의 이용거리와 같은 이용거리의 다른 데이터들에 대한 평균 이용시간을 확인해 보려고 한다.

```python
df[df['이용시간'].isnull()]
```

> 이용시간이 결측치인 모든 데이터 추출

```python
df[df['이용거리']==2290]
```

> 이용거리가 2290인 값을 출력

```python
df_group = df.groupby('이용거리')
df_group.mean()
```

> 이용거리를 기준으로 그룹화 시키고 그것에 따른 평균값을 구했다. group에 대한 자세한 내용은 추후 진행할 예정이다.

```python
print(df_group.mean().loc[2290])
df_group.mean().loc[2290]['이용시간']
```

> loc를 이용하면 row 정보인 index 정보를 얻어 올 수 있다. 입력되는 데이터는 문자열인 경우 '(홑따옴표)를 이용하며 숫자는 생략한다.
이렇게 얻은 정보에서 특정 feature를 얻으려면 직접 입력해 주면 된다.

#3. 데이터 중복 확인
> Dataframe에서는 중복 확인을 위해 duplicated라는 함수를 지원한다. 중복된 데이터가 나타나면 true를 반환한다.

```python
print(df['이용거리'][:5])
print(df.duplicated(['이용거리'])[:5])
```

0번이 False이고 1번이 True이다. 이런경우 1번이 중복된 데이터를 의미한다.

```python
print(df)
print(df.duplicated(['이용거리']))
df.duplicated(['이용거리', '나이'])
```

>feature가 두 개 이상 사용되면 and 연산으로 두 조건이 모두 만족되는 경우만 중복으로 처리한다.

```python
df.duplicated(['이용거리'], keep='last')
```

> keep의 기본값은 first이다. last 옵션을 주면 중복된 데이터가 나올 경우 마지막 값을 false로 변경해 준다.

```python
#결측치인 값이 존재하면 true로 반환
df.isnull()
```

```python
#feature별 결측치 집계
df.isnull().sum()
```

```python
def null_feature_indexs():
  features = df.isnull().sum()

  #결측치가 존재하는 feature의 index정보 추출
  null_features = features[features>0]
  return null_features.index.to_list() #리스트로 반환`
```

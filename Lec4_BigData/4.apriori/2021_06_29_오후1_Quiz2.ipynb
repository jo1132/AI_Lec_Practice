> 다음 내용을 apriori를 이용하여 분석하고 보고서를 작성하시오

```python
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
dataset = [ 
           ['아메리카노', '카페라떼'], 
           ['카페라떼', '아메리카노', '카푸치노'], 
           ['바닐라라떼', '아메리카노'], 
           ['녹차라떼', '카페라떼', '아메리카노'], 
           ['카페모카', '아메리카노'], 
           ['아메리카노', '카페라떼'], 
           ['초콜릿라떼', '아메리카노'], 
           ['아메리카노'], 
           ['카페모카', '카페라떼'] 
           ]
dataset
```

##1.1 데이터 변환

```python
te = TransactionEncoder()
te_fit = te.fit(dataset)
```

```python
arrMatrix = te_fit.transform(dataset)
arrMatrix
```

###2. 데이터 분석

```python
te_fit.columns_
```

```python
df = pd.DataFrame(data=arrMatrix, columns=te_fit.columns_)
df
```

```python
#2개 이상인 값이 너무 없어 0.1로
itemset = apriori(df, min_support=0.1, use_colnames=True)
itemset
```

> length피쳐 만들기

```python
itemset['Length'] = itemset['itemsets'].apply(lambda x : len(x))
itemset
```

```python
itemset_test = itemset[itemset['Length']>1]
itemset_test
```

```python
ar = association_rules(itemset, metric='confidence', min_threshold=0.3)
ar
```

#Quiz2

```python
dataset = [
 ['기저귀', '버터', '맥주'],
 ['기저귀', '맥주'],
 ['기저귀', '빵'],
 ['떡', '땅콩', '사이다']
]
dataset
```

##1. te 부터 df생성까지

```python
te = TransactionEncoder()
arr = te.fit(dataset).transform(dataset)
df = pd.DataFrame(arr, columns=te.columns_)
```

##2. 데이터 분석

```python
itemset = apriori(df, min_support=0.2, use_colnames=True)
ar = association_rules(itemset, metric='confidence', min_threshold=0.3)
ar
```

##강사님 코드

```python
def getDataFrame(dataset):
  te = TransactionEncoder()
  transform_matrix = te.fit(dataset).transform(dataset)
  return pd.DataFrame(transform_matrix, columns=te.columns_)
```

```python
dataset = [
 ['기저귀', '버터', '맥주'],
 ['기저귀', '맥주'],
 ['기저귀', '빵'],
 ['떡', '땅콩', '사이다']
]

df = getDataFrame(dataset)
df
```

```python
itemset = apriori(df, min_support=0.2, use_colnames=True)
ar = association_rules(itemset, metric='confidence', min_threshold=0.3)
ar
```

```python
ar.sort_values(['support', 'confidence', 'lift'], ascending=False).head()
```

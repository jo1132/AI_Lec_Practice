#Quiz1
> 다음 내용을 apriori를 이용하여 분석하고 보고서를 작성하시오

```python
dataset = [ 
           ['아메리카노', '카페라떼'], 
           ['카페라떼', '아메리카노', '카푸치노'], 
           ['바닐라라떼', '아메리카노'], 
           ['녹차라떼', '카페라떼', '아메리카노'], 
           ['카페모카', '아메리카노'], 
           ['아메리카노', '카페라떼'], 
           ['초콜릿', '아메리카노'], 
           ['아메리카노'], 
           ['카페모카', '카페라떼'] 
           ]
```

#1. 라이브러리 등록

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

#2. 데이터 변환

```python
te = TransactionEncoder()
arrayMatrix = te.fit(dataset).transform(dataset)
arrayMatrix
```

```python
import pandas as pd

df = pd.DataFrame(arrayMatrix, columns=te.columns_)
df
```

#3. 데이터 분석

```python
frequent_itemsets = apriori(df, min_support=0.2,  use_colnames=True)
frequent_itemsets
```

```python
ar = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.3)
ar.sort_values('lift', ascending=False)
```

> 최소 지지도를 0.2로 진행한 결과 상품들에 대한 연관성이 없는 것으로 나타난다. 
* 향상도 1 이하

> 따라서 최소 지지도를 0.1로 수정하여 다시 분석해 보자

```python
frequent_itemsets = apriori(df, min_support=0.1,  use_colnames=True)
ar = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.3)
ar.sort_values('lift', ascending=False)
```

> 지지도를 0.1로 변경한 후 실행한 결과 (녹차라떼)를  구매할 경우 (카페라떼, 아메리카노)를 구매할 확률이 높게 나타났다.
* 향상도 2.25, 신뢰도 1

> 위와 같이 향상도가 1 초과인 정보와 신뢰도가 높은 정보를 기준으로 추천해 주면 된다.

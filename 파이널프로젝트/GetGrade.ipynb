##이미지, 너비, 높이 입력받기

```python
import os

PATH = '/content/drive/MyDrive/Colab Notebooks/파이널프로젝트/data/농산물 품질(QC) 이미지/Training_unzip_miniset'
folder_list = os.listdir(PATH)
folder_list.sort()
folder_list
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from PIL import Image
from keras.preprocessing import image as Kimage

def GetImage(image_path):
  #테스트용이미지 불러오기
  img = Image.open(image_path)
  return img

def GetSize():
  width = float(input('너비 입력'))
  height = float(input('높이 입력'))

  pred = lasso.predict(pd.DataFrame(data=[width, height]).T)

  weight = pred[0]
  return (width, height, weight)

def GetFruit_Name(img):
  x = Kimage.img_to_array(img)
  x = x.reshape((1,)+x.shape)

  datagen = ImageDataGenerator(rescale=1./255)
  batch = datagen.flow(x, batch_size=1)
  pred = model.predict(batch)
  return pred.argmax()

def GetData(image_path):
  img = GetImage(image_path)
  width, height, weight = GetSize()
  fruit_name = GetFruit_Name(img)
  return (width, height, weight, folder_list[fruit_name])
```

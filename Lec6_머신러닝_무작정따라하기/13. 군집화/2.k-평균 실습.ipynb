#1. 데이터 읽기

```python
from sklearn.datasets import load_iris

iris = load_iris()
iris.feature_names
```

```python
import pandas as pd

df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df_iris.head(2)
```

#2. 모델 만들기

##2.1 군집화

```python
from sklearn.cluster import KMeans
# n_clusters는 몇 개로 분류할 것인지 선정하는 것이며 우리는 이미 iris가 3개의 품종으로 이루어 진것을 알고 있다.
# init은 사용할 알고리즘을 지정하는데 기존에 k-means는 중심점을 지정하는데 문제가 있어 이를 해결하기 위해 k-means++을 적용
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300)
kmeans.fit(df_iris)
# lables를 통한 결과 확인
kmeans.labels_
```

```python
iris.target
```

```python
df_iris['target'] = iris.target
df_iris['labels'] = kmeans.labels_
df_iris.groupby(['target', 'labels'])['sepal length (cm)'].count()
```

> target은 원본 데이터를 의미하며 labels는 군집화를 통한 분류된 결과이다. 원본 데이터의 0과 lables의 1 모두 50으로 나타났으며 분류가 잘 된 것을 알 수 있다.

> target의 1은 lables의 0과 2로 분류되어 있으며 48, 2로 2개 정도의 오차 범위가 발생하였다.

> target의 2는 차이가 많이 난다.

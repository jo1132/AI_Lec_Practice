# 강사님 코드

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
```

#2. 데이터 전처리

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
standard_scaler = scaler.fit_transform(cancer.data)
```

#3. 모델 만들기

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(standard_scaler, cancer.target, test_size=0.3, random_state=62)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```

#4. 교차 검증


##4.1 cross_val_score

```python
from sklearn.model_selection import cross_val_score
lr = LogisticRegression()
scores = cross_val_score(lr, standard_scaler, cancer.target, cv=5)
scores.mean()
```

##4.2 하이퍼 파라미터

```python
from sklearn.model_selection import GridSearchCV

def get_best_param(cost_list):
  params = {
      'penalty' : ['l2'],
      'C' : cost_list
  }

  grid_cv = GridSearchCV(lr, param_grid=params, cv=5, scoring='accuracy')
  grid_cv.fit(standard_scaler, cancer.target) 
  print('====================================================================')
  print('best score :', grid_cv.best_score_)
  print('bset params :', grid_cv.best_params_)
  print('====================================================================')
```

```python
C_list = [0.001, 0.01, 0.1, 1, 10, 25, 50, 100]
get_best_param(C_list)
```

> best score : 0.9806862288464524

> bset params : {'C': 1, 'penalty': 'l2'}

```python
C_list = list(range(1, 501))
for i in range(len(C_list)):
  C_list[i] = C_list[i]/100
get_best_param(C_list)
```

#7. 최적의 파라미터

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(penalty='l2', C=0.26)
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```

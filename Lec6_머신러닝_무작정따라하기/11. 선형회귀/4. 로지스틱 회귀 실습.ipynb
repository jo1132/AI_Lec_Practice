#1. 데이터 읽기

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
```

#2. 데이터 전처리

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
standard_scaleed=scaler.fit_transform(cancer.data)
```

#3. 모델 만들기

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(standard_scaleed, cancer.target, test_size=0.3, random_state =62)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```

#4. 교차 검증

##4.1 cross_val_score

```python
from sklearn.model_selection import cross_val_score
lr = LogisticRegression()
scroes = cross_val_score(lr, standard_scaleed, cancer.target, cv=5)
scroes.mean()
```

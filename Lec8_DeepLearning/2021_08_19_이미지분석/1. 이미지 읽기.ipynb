#2. 동영상 불러오기

##2.1 동영상 정보 확인

```python
import cv2

movie_path = '/content/drive/MyDrive/Colab Notebooks/Lec8_DeepLearning/2021_08_19_이미지분석/mov/mov01.avi'
mov = cv2.VideoCapture(movie_path)

width = mov.get(cv2.CAP_PROP_FRAME_WIDTH)
height = mov.get(cv2.CAP_PROP_FRAME_HEIGHT)
count = mov.get(cv2.CAP_PROP_FRAME_COUNT)
fps = mov.get(cv2.CAP_PROP_FPS)

print(width, ' : ', height)
print('총 frame 수 : ', count)
print('초당 frame 수 : ', fps)
```

#3. 동영상을 이미지로 저장하기

```python
import cv2

path = '/content/drive/MyDrive/Colab Notebooks/Lec8_DeepLearning/2021_08_19_이미지분석/'
movie_path = path + 'mov/mov01.avi'
mov = cv2.VideoCapture(movie_path)

for i in range(10):
  mov.isOpened()
  isRead, frame = mov.read()
  print(isRead)
  if isRead:
    print('readed')
    file_path = path + 'snapshot/snapshot_'+str(i)+'.jpg'
    cv2.imwrite(file_path, frame)

mov.release()
cv2.destroyAllWindows()
```

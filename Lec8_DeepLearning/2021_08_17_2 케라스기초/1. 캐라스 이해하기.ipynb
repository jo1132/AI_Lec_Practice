#1. 캐라스란
> 캐라스는 딥러닝 모델을 만들기 위한 고수준의 굿어 요소를 제공하는 모델 수준의 라이브러리이다. 

image.png

> 케라스는 백엔드 엔진 기반에 동작되기 때문에 쉽고 간단하게 학습할 수 있는 딥러닝 라이브러리 이다.
- TensorFlow : 구굴에서 개발
- Theano : 몬트리올 대학 MILA 연구소에서 개발
- CNTK : 마이크로소프트에서 개발

>  CUDA를 이용하여 GPU로 데이터를 처리할 수 있다.

#2. GPU vs CPU

image.png

> GPU는 그래픽 처리 장치로 그래픽 연산에 특화된 부품이다. 원리 자체만 놓고 보면 CPU와 같지만, CPU가 복잡한 연산을 처리하기 위해 연산 속도에 집중한 반면, GPU는 '쉬운 작업을 대량'으로 처리한다는 것이 차이점이다

#1. 경사 하강법
> 오차가 가장 적은 위치를 선정하기 위한 알고리즘이다. 산 정상에서 가장 낮은 위치로 이동하는 방법으로 내려오는 것을 비유하여 경사 하강법이라고 표기되었다고 생각된다.

##1.1 이해하기

1_ZmzSnV6xluGa42wtU7KYVA.gif

> 위의 그림을 보면 왼쪽의 가장 높은 지역에서 낮은 쪽으로 이동하기 위해 x값을 증가하는 것을 볼 수 있다. 경사 하강법이란 이렇게 가장 낮은 위치로 이동하는 알고리즘을 의미한다.

##1.2 local minima 문제

image.png

> 위의 그림에서 보듯이 가장 낮은 지점은 오른쪽에 나타나는 Global minimum이지만 왼쪽과 같이 특정 지점에서만 작은 지점을 찾게 되어 가장 낮은 지점을 못찾는 문제가 발생한다.

#2. 확률적 경사 하강법
> 위와 같은 문제를 해결하기 위해 여러 알고리즘이 나타나게 된다.

image.png

> 위의 그림에서 보듯이 기본 베이스는 GD(Gradient Descent)이다. 모든 내용을 다 확인하고 움직이기에는 시간이 오래 걸리니 이를 줄여서 SGD가 나오게 되었고 스텝의 방향이냐 사이즈냐에 따라 여러 알고리즘이 나왔으며 적절한 조절을하는 Adam이 가장 많이 사용되기는 한다. 하지만 가장 좋은 것이 아님으로 상황에 따라 여러 알고리즘들을 대입하여 사용해야 한다.

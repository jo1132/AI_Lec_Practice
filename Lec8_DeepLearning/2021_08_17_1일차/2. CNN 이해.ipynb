#4.2 ReLU(Rectified Linear Unit)
> 선(Linear)의 모양을 변경하는(Rectified) 정도로 해석하면 될듯 하다. ReLU 함수는 sigmoid가 특정 수치를 넘어가면 미분값이 0으로 수렴하는 문제를 해결하기 위해 만든 함수이다.

image.png

> 위의 그림에서 보듯이 음수인 경우 0으로 표현하고 양수인 경우 y=x로 표현하여 이미지 분석이 주로 사용된다.

#5. 손실함수
> 훈련 데이터에서 신경망의 성능을 측정하는 방법으로 신경망네트워크가 옳은 방향으로 학습될 수 있도록 도와준다.
- binary_crossentropy : 이진 분류 손실 함수로 2개의 label이 나타날 경우 사용된다.
- categorical_crossentropy : 다중 분류 손실 함수로 ont-hot encoding 결과를 얻게 된다.
- sparse_categorical_crossentropy : 다중 분류 손실 함수로 integer type으로 표현된다.

#1. 텐서(Tensor)
- 벡터 데이터 : (samples, features)로 구성된 2D 텐서로 대부분의 데이터가 여기에 속한다.
- 시계열 데이터 또는 시퀀스 데이터 : (samples, timesteps, features)로 구성된 3D텐서로 시간이 중요한 feature일 경우 사용된다. 주식 데이터셋이 여기에 포함된다.
- 이미지 : 텐서플로우 방식(samples, height, width, channels) 또는 씨아노 방식(samples, channels, height, width)로 구성된 4D 텐서
- 동영상 : (samples, frames, height, width, channels) 또는 (samples, frames, channels, height, width)로 구성된 5D 텐서

#2. 배치 데이터
> batch는 네이버 영어 사전에서 "(일괄 처리를 위해) 함께 묶다"로 표현된다. 컴퓨터에서는 데이터 처리를 한 번에 모아서 처리할 때 batch라고 표현한다.

> 위의 텐서에서 보듯이 첫 데이터는 sample들을 의미한다. 이 샘플을 일괄적으로 묶어서 처리할 경우 사용된다. 많은 sample을 한번에 처리하기 힘들기 때문에 분할해서 처리하게 된다.

> 이런 배치 데이터를 다룰 때 첫 번째 축(samples 축)을 배치 축 또는 배치 차원이라고 부른다.

#3. 신경망 구조

image.png

> 신경망은 위와 같은 구조를 갖는다. 데이터가 들어오면 가중치와 연산하여 데이터를 변환시킨다. 다음 레이어로 이동하면서 다시 다중치와 연산하여 예측값을 구한다. 이 예측값과 실 데이터를 손실 함수에 넣고 손실 점수를 구한다. Optimizer에서는 손실 점수를 줄이기 위해 가중치를 수정하고 다시 데이터 변환를 진행하게 된다. 

##3.1 Layer
> Layer는 하나 이상의 텐서를 입력 받아 하나 이상의 텐서를 출력하는 데이터 처리 모듈이다.
- 밀집 연결 Layer(Densely Connected Layer) : 벡터 데이터일 경우 사용되며 완전 연결 Layer(Fully Connected Layer)라고 하기도 한다. 캐라스에서는 Dense 클래스를 사용한다.
- 순환 Layer(Recurrent layer) : 시퀀스 데이터일 경우 사용한다.
- 2D 합성곱 Layer(Convolution Layer) 이미지 데이터일 경우 사용된다. 캐라스에서는 Conv2D를 사용한다.

### (실습, item_name, choice_description, new_price로 구성된 품목별 단가 chip_price df를 만들기

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv", sep="\t")

df["new_price"] = df['item_price'].str[1:-1]   #new_price만들기

#price가 10달러 이상이고 , quantity가 1개인항목만 따로 고르기
df2 = df[(df["new_price"].astype(float) > 10) & (df['quantity'] == 1)]

#중복 제거하기
df2 = df2.drop_duplicates(['item_name'])

#df2에서  상품명, 초이스 뭐시깽이, 가격으로만 구성된 df만들기
chip_price = df2[["item_name", 'choice_description', 'new_price']]

#chip_price.reset_index() #인덱스 다시생성
chip_price = chip_price.reset_index(drop=True) #기존인덱스는 삭제
chip_price
```

### (Chip_price에서 item_name순으로 sort) ###


```python
#인덱스로 정렬할땐 : .sort_index()

#열로 정렬할땐
chip_price = chip_price.sort_values(by= "item_name").reset_index(drop=True)
chip_price
```

### 가장 비싼 품목(제품)은? ###

```python
chip_price.sort_values(by="new_price", ascending=False).head()
```

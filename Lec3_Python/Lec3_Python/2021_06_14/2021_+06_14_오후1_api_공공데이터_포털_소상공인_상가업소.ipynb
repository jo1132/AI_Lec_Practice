```python
end_point = 	"http://apis.data.go.kr/B553077/api/open/sdsc/"
```

```python
op_name = "storeZoneInRadius?radius=500&cx=127.004528&cy=37.567538&ServiceKey="
```

```python
data_form = "&type=json"
```

```python
url = end_point + op_name + my_key + data_form
print(url)
```

```python
import urllib.request
res = urllib.request.urlopen(url)
res_str = res.read().decode('utf-8')

#str 형식을 json으로 변환
import json #파이썬 내장 함수
res_json = json.loads(res_str) #json형태대로 만들어짐 -> dictionary 형태랑 비슷하게됨
print(type(res_json)) # 타입이 dict가 되었다.

#json 에서 pandas 로 바꾸기
#from pandas.io.json import json_nomalize
import pandas as pd

df = pd.json_normalize(res_json['body']['items'])
```

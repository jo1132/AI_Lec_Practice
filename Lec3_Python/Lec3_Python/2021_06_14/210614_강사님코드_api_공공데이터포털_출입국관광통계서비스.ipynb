### ( 실습. 출입국 관공통계 수집후 데이터 프레임 만들기  ) 
1. 국가 3개 -> 112(중국), 일본(130), 미국(275)
2. 기간 -> 2019, 2020년도 1월~12월  
3. ym(년월), natKorNm(국가), num(방문객수), edCD(구분) 
4. 년월  국가  방문객수  구분 

```python
import urllib.request
import json

def get_visit(url):
    res = urllib.request.urlopen(url)
    json_str = res.read().decode("utf-8")
    json_object = json.loads(json_str) 
    item = json_object["response"]["body"]["items"]
    return item
```

```python
end_point1 = "http://openapi.tour.go.kr/openapi/service/EdrcntTourismStatsService/"
end_point2 = "getEdrcntTourismStatsList?"
parm_key = "_type=json&serviceKey=" + my_key

country_code = ["112","130","275"]
visit_year = [2019, 2020]
ed_cd = "E"

init = 0 
for country in country_code:
    for year in visit_year:
        for month in range(1,13):
            parm_ym = "&YM=" + str(year) + "%02d"%month
            parm_nat_cd = "&NAT_CD=" + country
            parm_ed_cd = "&ED_CD=" + ed_cd 
            url = end_point1 + end_point2 + parm_key
            url = url + parm_ym + parm_nat_cd + parm_ed_cd 
            print(url) ; print("-"*60)
            result = get_visit(url) 
            df = pd.DataFrame([result])
            if init == 0:
                df_total = df.copy() ;init = 1
            else:
                df_total = pd.concat([df_total, df])
```

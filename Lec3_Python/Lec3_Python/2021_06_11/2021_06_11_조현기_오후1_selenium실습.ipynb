```python
def Video(item):
    #text정보있는 div가져오기
    text_wrap = item.find("div", {"class":"text-wrapper"})
    
    #채널 가져오기
    channel = text_wrap.find("ytd-channel-name", {"id":"channel-name"}).find('a').text
    
    #title가져오기
    title_A = text_wrap.find("div", {"id":"title-wrapper"}).find('a')
    title = title_A["title"]
    
    #비디오 Url 가져오기
    video_url = "https://www.youtube.com/" + title_A['href']
    
    #view count
    spans = text_wrap.find_all("span", {"class" : "ytd-video-meta-block"})
    view = spans[0].text

    #upload_data
    upload = spans[1].text
    
    return [channel, title, video_url, view, upload]
```

```python
def Playlist(item):
    #text정보있는 div가져오기
    content = item.find("div", {"id":"content"})
    
    #채널 가져오기
    channel = content.find("ytd-channel-name", {"id":"channel-name"}).find('a').text

    #title가져오기
    title = content.find("span", {"id" : "video-title"})

    #플레이링크가져오기
    link_A = content.find('a')
    link_url = "https://www.youtube.com/" + link_A["href"]
    
    #플레이 리스트 가져오기
    ul_list = content.find_all("li")
    list1 = ul_list[0].find("yt-formatted-string", {"id" : "video-title"}).text
    
    list2 = ul_list[1].find("yt-formatted-string", {"id" : "video-title"}).text
    
    return [channel, title, link_url, list1, list2]
```

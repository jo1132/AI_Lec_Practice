###**클래스(Class)**
명령이 모여 함수가 되고, 비슷한 기능의 함수들을 모아 클래스를 만든다.
클래스에 있는 기능을 변수로 가져와서 생성한다.
그것을 객체라고도 하고, 인스턴스라고도 한다.

class 클래스명:
  클래스 변수 선언
  함수....
  함수....

```python
#사용자가 건들지 못하게 한다.
class Calculator:
  txt = "문자"
  def sum(a, b):
    c = a + b
    return c
```

```python
type (Calculator())
```

```python
#새로운 인스턴스 생성, 객체 할당....
x = Calculator
```

###**클래스 상속**

```python
class Family():
  def father(self):
    print("아빠")
  def mother(self):
    print("엄마")
  def son(self):
    print("아들")
class Look():
  def hansom(self):
    print("잘 생겼다.")
  def pretty(self):
    print("예쁘다")
  def slim(self):
    print("날씬하다")
```

```python
a = Family()
b = Look()

a.father()
b.hansom()

a.mother()
b.slim()

a.son()
b.pretty()
```

4. 화면에서 입력받아 처리


```python
def conv_fahr(x):
  return (9.0/5.0)*x + 32
```

```python

x = float(input("섭씨 온도는? "))
y = conv_fahr(x)
print("섭씨 %s는 화씨 %3.2f 입니다." %(x, y))
```

**(아래 섭씨 데이터를 화씨로 변환하기)**

```python
x = [25, 20, 32, 77]
for i in x:
  print("섭씨 %s는 화씨 %3.2f 입니다." %(i, conv_fahr(i)))
```

```python
[conv_fahr(i) for i in x]
```

###**(화면에서 섭씨를 입력받아 반복 처리문 만들기)
1. while(True)
2. 입력이 0이면 종료


```python
while True:
  C = float(input("섭씨 온도는? "))
  if C == 0:
    print("종료합니다.")
    break
  print("화씨: ",conv_fahr(C))
```

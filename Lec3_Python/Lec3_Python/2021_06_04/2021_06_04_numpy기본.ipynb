###**행렬연산**

```python
arr1 = np.array([[1, 2],[3, 4]])
arr2 = np.array([[5, 6],[7, 8]])

print(arr1 + arr2)
np.add(arr1, arr2)
```

```python
print(arr1*arr2)
np.multiply(arr1, arr2)
```

```python
#행렬 곱  행렬곱에서 
#B의 열의 수가 A의 행의 수와 맞아야한다.
#A(3, 5) => B(5, 8)
print(arr1.dot(arr2))

np.dot(arr1, arr2)
```

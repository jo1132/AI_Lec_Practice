###**반복문 조건문 복습**

for 변수명  in (iterable):
  실행문

while(조건문):
  실행문
  조건 변경 행위(반복에서 탈출하기 위한)


```python
i = 0
while(i<6):
  print(i)
  i += 1
```

```python
i = 0
while(i<6):
  print(i)
  if i == 3:
    break
  i += 1
```

```python
selected = None
while selected not in ["가위", "바위", "보"]:
  selected = input("가위, 바위, 보 중 선택하세요")
print("선택 = ", selected)
```

```python
import random
correct_answer = random.randint(1, 5)

count = 0

while True:
  number = int(input("숫자를 입력하셍"))
  count += 1
  if correct_answer == number:
    break
  elif correct_answer > number:
    print(number, " 보다 놉다")
  else:
    print(number, " 보다 작다")
print("정답이야 축하행", count, "번째 시도만에 성공")
```

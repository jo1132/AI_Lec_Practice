```python
#3 보다 크면 제곱한 수를 만들기
num = list(range(1, 7))
square = []
for i in num:
  if i >= 3:
    square.append(i**2)
square
```

```python
square = [i**2 for i in num if i>= 3]
square
```

###**(type 3)**
1. 1~6으로 구성된 num
2. 3보다 크면 제곱을 하고
3. 3보다 작으면 3제곱을 한 리스트 만들기

```python
num = list(range(1, 7))
bb = []
for i in num:
  if i > 3:
    bb.append(i**2)
  else:
    bb.append(i**3)
bb
```

```python
cc = [i**2 if i>3 else i**3 for i in num]
cc
```

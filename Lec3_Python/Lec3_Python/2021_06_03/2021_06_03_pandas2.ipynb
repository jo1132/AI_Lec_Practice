```python
#열번쨰, 여러 컬럼 추출할때

df = german[["Creditability", "Account Balance","Duration of Credit (month)" ]]
df
```

```python
#열 한번쨰, 
for row in range(0, len(df), 100):
  print("row = ", row)

  for col in range(0, df.shape[1]):
    print("row = ", row)
    print("col = ", col)
    print("value = ", df.iloc[row.col])
    print("="*50)
  break
```

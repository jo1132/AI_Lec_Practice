###**Serise 연산**

```python
#Serise 연산
fruit1 = pd.Series(data=[5, 6,7, 8], index=['ap','ba', 'ch', 'pe'])

fruit2 = pd.Series(data=[3, 2, 9, 5, 10], index=['ap','or', 'na', 'ch', 'pe'])

print(fruit1)
print(fruit2)
```

```python
fruit1+fruit2
# 서로 같은것만 더해지고 왠지 float형이 되었다.
```

```python
#데이터 프레임 연산
fd1 = {'Ohio':[4, 8, 3, 5], 'Texas':[0, 1, 2, 3]}
fruitFrame1 = DataFrame(fd1, columns=['Ohio', 'Texas'], index=['apple','banana', 'cherry', 'peer'])
fd2 = {'Ohio':[3, 0, 2, 1, 7], 'Colorado':[5, 4, 3, 6, 0]}
fruitFrame2 = DataFrame(fd2, columns=['Ohio', 'Colorado'], index=['apple','orange','banana', 'cherry', 'mango'])
print(fd1)
print(fd2)
```

```python
fruitFrame1 + fruitFrame2  #중복되는 항목들만 합쳐진다. 그리고 왠지 더블형이다.
```

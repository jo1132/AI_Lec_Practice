###**(실습)**
딕셔너리에서 value가 78인 key를 찾아서 출력(key = ???)

```python
d = {
    'height' : 180,
     'weight' : 78,
     'temp' : 36,
     'eye' : 1
}

print([ i for i in d.items() if i[1] == 78])

x = [ i for i in d.items() if i[1] == 78]
print(type(x))
x = x[0]
print(type(x))
```

#**(실습)** 딕셔너리에서 value가 78인 key를 찾아서 출력 (key = ??)


```python
d = {
    '1' : [60, 80, 30],
     '2' : [78,54, 98],
     '3' : [100,50, 70],
     '4' : [70, 60, 80]
}
print("key = ",[ i[0] for i in d.items() if (78 in i[1])])
```

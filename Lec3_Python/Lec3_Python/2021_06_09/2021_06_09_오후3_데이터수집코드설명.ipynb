###**(실습, 네이버 뉴스 속부 수집)**
1. 데이터는 다 가지고 왔음
  - 이미지 주소 img_url
  - 기사 주소 link_content_url
  - 기사 제목  link_title
  - 기사 내용 content
2. 데이터로 만들어서 
  - 데이터 프레임만듬
3. 컬럼은 순서대로하면 됨

```python
for item2 in dl:
  try:
    img = item2.find("dt", {"class" : ""}).find("a")
    print(img["href"])
  except:
    print("No")
```

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

tot_list = []
for item2 in dl:
  try:
    img = item2.find("dt", {"class" : "photo"}).find("img")
    img_url = img["src"]
  except:
    img_url = np.nan

  try:
    img = item2.find("dt", {"class" : ""}).find("a")
    link_content_url = img["href"]
  except:
    link_content_url = np.nan

  try:
    img = item2.find("dt", {"class" : ""}).text.replace("\t", "").replace("\n", "").replace("\r", "")
    link_title = img
  except:
    link_title = np.nan
  try:
    img = item2.find("span", {"class" : "lede"}).text
    content = img
  except:
    content = np.nan

  tot_list.append([img_url, link_content_url, link_title, content])
tot_list
```

```python
df = pd.DataFrame(data=tot_list, columns=['img_url', 'link_content_url', 'link_title', 'content'])
df
```

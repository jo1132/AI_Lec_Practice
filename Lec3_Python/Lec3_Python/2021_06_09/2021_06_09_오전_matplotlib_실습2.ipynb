```python
def my_plotter(ax, data1, data2, param_dict):
  out = ax.plot(data1, data2, **param_dict)
  return out
```

###**(Pie차트 +인자 실습)**

```python
tips.columns
```

```python
def drawpie(col):
  labels = tips[col].value_counts().index
  nums = tips[col].value_counts().values


  fig, ax = plt.subplots()
  ax.pie(nums, labels=labels, autopct="%1.2f%%", startangle=90)
  ax.axis("equal")
  ax.set_title(col)
  plt.show()
```

```python
drawpie("day")  #column 을 입력하면 그에따른  pie그래프가 생성된다.
```

###**(요일별로 total_bill평균 분포를 pie챠트로 그리기)**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips.head(3)
```

```python
group = tips.groupby("day").mean()
group
```

```python
labels = group.index
labels
```

```python
datas = group.total_bill.values
datas
```

```python
fig, ax = plt.subplots()
explode = (0, 0, 0, 0.1)
ax.pie(datas, explode=explode, labels=labels, startangle=90, autopct = "%.2f%%", shadow=True)
ax.set_title("total_bills per day")
ax.axis('equal')
plt.show()
```

```python
#강사님 코드
def day_pie(values, index):
  fig, ax = plt.subplots()
  ax.pie(values, labels=index, startangle=90, autopct = "%d $", shadow=True)
  ax.axis('equal')
```

```python
group = tips.total_bill.groupby(tips["day"]).mean()
tips_index = group.index
tips_value = group.values
day_pie(tips_value, tips_index)
```

###**(흡연자의 요일별 total_bill 평균분포 그리기**

```python
tips.smoker.value_counts()
```

```python
#위에서 정의한 def day_pie사용해도 될것같
smoker = tips[tips["smoker"] == 'Yes']
smoker
```

```python
total_bill = smoker.groupby("day").std().total_bill
total_bill
```

```python
tips_index = total_bill.index
tips_value = total_bill.values
day_pie(tips_value, tips_index)
```

###**(실습. 년도를 입력받아 윤년인지 아닌지 판별하는 함수(leap_year)를 작성**
1.화면에서 연도를 입력
2. 윤년이면 "윤년" 아니면 "평년"을 출력
3. 윤년은 (2월이 29일까지)
  - 4로 나눠서 떨어지면 윤년
  - 4, 100으로 나눠서 떨어지면 평년
  - 4, 100, 400으로 나눠서 떨어지면 윤년

```python
def leap_year(year):
  ch = "평"
  if year % 4 == 0:
    ch = "윤"
    if year % 100 == 0:
      ch = "평"
      if year % 400 == 0:
        ch = "윤"
  return ch

def leap_year2(year):
  ch = False
  if year % 4 == 0:
    ch = True
    if year % 100 == 0:
      ch = False
      if year % 400 == 0:
        ch = True
  return ch
```

```python
years = [1900, 1988, 2000, 2008, 2020, 2100, 2400]
for i in years:
  print(i, "년은 ", leap_year(i), "년 입니다.", sep="")

print(list(map(leap_year, years)))

print(years,"중에서 윤년은", list(filter(leap_year2, years)), "입니다.")
```

```python
list(map((lambda year : "윤년" if ((year%4==0) and (year%100 != 0) or (year%400 == 0)) else "평년"), years))
```

###**(반복처리)**
1. for()함수 leap_year() 사용해서
2. aa의 결과값을 y로 만들기

```python
aa = [1980, 2019, 2020]

for year in aa:
  print("year = ", leap_year(year), "년", sep="")
```

###**(while)**
1. 화면에서 연도를 입력받아 윤년, 평년 판별
2. 화면에서 0이면 while문 종려료

```python
y = []
while True:
  year = int(input("연도를 입력하세요"))
  if year == 0:
    break;
  result = leap_year(year)
  y.append([year, result])
  print(year,"년도는 ", result, "년 입니다.\n", sep="")

print(y)
```

###**(함수 실습, 입력자료에서 홀수, 짝수, 문자열 숫자세는 함수 만들기)**
1. 1, 2, 3, "ab", 6, 7 -> 홀수 3, 짝수 2, 문자 1
2. 입력자료는 몇개가 들어올지 모른다.


```python
def filter_func(item):
  if type(item) == str:
    return "문자"
  elif item%2 == 0:
    return "짝수"
  else:
    return "홀수"
```

```python
items = [1, 2, 3, "ab", 6, 7]
list(map(filter_func, items))
```

```python
def count_input(*args):
  y = {
      "문자" : 0,
      "짝수" : 0,
      "홀수" : 0
  }
  ch =""
  for item in args:
    if type(item) == str:
      ch  = "문자"
    elif item%2 == 0:
      ch = "짝수"
    else:
      ch = "홀수"
    y[ch] += 1
    print(ch)
  return y
```

```python
print(count_input(1, 2, "a"))
```

```python
df = pd.DataFrame(data=count_input(1, 2, "a"), index=["item"])
```

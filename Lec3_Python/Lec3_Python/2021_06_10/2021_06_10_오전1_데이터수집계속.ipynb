```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

###**(함수만들기)**
1. get_page() -> url을 입력받아서 dl화일 리턴
2. page_to_items() -> dl을 입력받아서 news_page list를 return
3. news_page로 DataFrame 만들기
4. DataFrame에서  csv만들기

```python
def get_page(url):
  header = {"User-Agent":"Mozilla.5.0"}
  html = requests.get(url, headers = header).text
  soup = BeautifulSoup(html, "html.parser")
  type06 = soup.find("ul", {"class" : "type06_headline"})
  return type06.find_all("dl")  #list 형으로 가져옴
```

```python
def page_to_items(dl):
  tot_list = []
  for item2 in dl:
    #이미지 소스 가져오기
    try:  
      img = item2.find("dt", {"class" : "photo"}).find("img")
      img_url = img["src"]
    except:
      img_url = np.nan
    #뉴스 주소 가져오기
    try:
      img = item2.find("dt", {"class" : ""}).find("a")
      link_content_url = img["href"]
    except:
      link_content_url = np.nan
    #뉴스 타이틀 가져오기
    try:
      img = item2.find("dt", {"class" : ""}).text.replace("\t", "").replace("\n", "").replace("\r", "")
      link_title = img
    except:
      link_title = np.nan
    #뉴스 기사 요약 가져오기
    try:
      img = item2.find("span", {"class" : "lede"}).text
      content = img
    except:
      content = np.nan
    tot_list.append([img_url, link_content_url, link_title, content])
  return tot_list
```

```python
url = input("url을 입력하세요: ")
dl = get_page(url)  #dl가져오기
data_list = page_to_items(dl) #읽어온 정보 리스트로
df = pd.DataFrame(data=data_list, columns=["이미지 주소", "링크 주소", "타이틀", "기사내용"])
df.to_csv("/content/drive/MyDrive/Colab Notebooks/2021_06_10/naver_news1.csv", index = False)
readdf =  pd.read_csv("/content/drive/MyDrive/Colab Notebooks/2021_06_10/naver_news1.csv")
readdf
```

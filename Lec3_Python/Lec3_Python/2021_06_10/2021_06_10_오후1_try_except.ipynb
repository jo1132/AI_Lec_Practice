```python
def divide(x, y):
  try:
    result = x/y
  except ZeroDivisionError:
    print("0으로 나눌 수 없음!!!")
  except TypeError:
    print("Type Error")
  else:
    print(x, "/", y, "= ", result)
  finally:
    print("마지막 finally는 무조건 실행!")

for i in range(5, -1, -1):
  divide(10, i)

divide("10", "0")
```

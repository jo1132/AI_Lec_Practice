###**Set Methods**
set은 순서가 중요하지 않다.

add(): Adds an elements to the set

```python
fruits = {"apple", "banana", "cherry"}

fruits.add("orange")
print(fruits)
```

```python
fruits = {"apple", "banana", "cherry"}  # add하면 중복된값은 입력되지 않는다.
fruits.add("apple")
print(fruits)
```

clear(): Removes all the elements from the set

```python
fruits = {"apple", "banana", "cherry"}

fruits.clear()

print(fruits)
```

copy(): Returns a copy of the set

```python
fruits = {"apple", "banana", "cherry"}

x = fruits.copy()

print(x)
```

difference(): Return a set containing the difference between two or more sets
x에만 존재하고 y에는 존재하지 않는 값을 리턴해준다.

```python
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}

z = x.difference(y)
print(z)

z = y.difference(x)
print(z)
```

difference_update(): Remove the items in this set that are also included in anoter, specified set
그냥 difference와 다르게 중복되는 원소를 삭제해준다.


```python
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}
x.difference_update(y)

print(x)
```

discard(): Remove the specified item 셋에서 아이템 삭제
discard가 remove와 다른 이유는 remove서는 원소가 없다면 에러를 발생시키지만 discard는 아니다


```python
fruits = {"apple", "banana", "cherry"}
fruits.discard("banana")
print(fruits)
```

intersection(): Returns a set, that is the intersection of two othersets
diffrence와 다르게 intersection은 중복되는 원소만 뽑아준다.
set.intersection(set1, set2 ... etc)

```python
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}

z = x.intersection(y)
print(z)
```

```python
x = {"a", "b", "c"}
y = {"c", "d", "e"}
z = {"f", "g", "c"}

result = x.intersection(y, z)   ## 두개 이상의 set집합도 비교 가능하다.
print(result)
```

intersection_update(): Removes the items in this set that are not present in other, specified set(s) 중복되는 원소만 남기고 전부 삭제해준다.

```python
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}

x.intersection_update(y)
print(x)
```

```python
x = {"a", "b", "c"}
y = {"c", "d", "e"}
z = {"f", "g", "c"}

x.intersection_update(y, z)   ##마찬가지로 두개 이상의 set에서도 가능하다

print(x)
```

isdisjoint(): Returns whether two sets have a intersection or not
중복이 있는지 없는지 확인해줌
중복이 없으면 참

```python
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "facebook"}
z = {"apple"}

result = x.isdisjoint(y)
print(result)

result = x.isdisjoint(z)
print(result)
```

issubset(): Returns whether this set contains anoter set or not
포함관계인지 아닌지 확인

```python
x = {"a", "b", "c"}
y = {"f", "e", "d", "c", "b", "a"}

z = x.issubset(y)
print(z)
```

```python
x = {"a", "b", "c"}
y = {"f", "e", "d", "c", "b"}

z = x.issubset(y)

print(z)
```

issuperset(): Returns whether this set contains anther set or not
서브셋과 반대로 내가 누구를 포함하는지 확인해줌

```python
x = {"f", "e", "d", "c", "b", "a"}
y = {"a", "b", "c"}
z = {'k', 'g', 'q'}

result = x.issuperset(y)
print(result)

result = x.issuperset(z)
print(result)
```

pop(): Removes an element from the set
랜덤으로 원소가 정해져 삭제되고 삭제된 원소를 반환해준다.

```python
fruits = {"apple", "banana", "cherry"}

fruits.pop()
print(fruits)
```

remove(): Removes the specified element

```python
fruits = {"apple", "banana", "cherry"}

fruits.remove("banana")
print(fruits)
```

symmetric_difference(): Return a set with the symmetric differences of two sets
중복되지 않은 원소들을 하나의 세트로 만들어 반환한다.

```python
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}

z = x.symmetric_difference(y)

print(z)
```

sysmetric_difference_update(): inserts the symmetric differences from this set and anoter 중복되는 원소 빼고 중복안되는 원소만 넣어준다.

```python
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}
x.symmetric_difference_update(y)

print(x)
print(y)
```

union(): Return a set containing the union of sets 원소를 중복없이 합쳐서 반환해준다.

```python
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}
z = x.union(y)
print(z)
```

```python
x = {"a", "b", "c"}
y = {"f", "d", "a"}
z = {"c", "d", "e"}
a = [1, 2, 3]   #set뿐만 아니라 iterable인것은 다 넣어준다.

result = x.union(y, z, a)   #순서상관없이 중복원소 빼고 set로 다 넣어준다.
print(result)
```

update(): Update the set with the union of this set and others
union은 새로운 set를 반환하지만 update는 자기 자신을 바꾼다.?

```python
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}
x.update(y)
z = x.union(y)
print(x)
print(z)
```

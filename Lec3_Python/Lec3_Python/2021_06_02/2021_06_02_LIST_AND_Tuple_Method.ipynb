###**리스트 메소드**

append(): Adds an element at the end of the list

```python
fruit = ['appel', 'banana', 'cherry']
fruit.append("orange")
print(fruit)
```

```python
fruit.append(fruit[2])
print(fruit)
```

clear(): Removes all the elements from the list


```python
fruit = ['apple', 'banana', 'cherry', 'orange']
print(fruit)
fruit.clear()
print(fruit)
```

copy(): Returns a copy of list


```python
fruits = ['apple', 'banana', 'cherry', 'orange']
x = fruits.copy()
print(x)
```

count(): Returns the number of elements with the specified value

```python
fruits = ['apple', 'banana', 'cherry', 'cherry']
x = fruits.count("cherry")
print(x)
```

extentd(): Add the elemtents of a list(or ant iterable), to the end of the current list
끝에 리스트, 세트, 튜플등 원소를 추가해중다.

```python
fruits = ['apple', 'banana', 'cherry']
cars = ['Ford', 'BMW', 'Volvo']
fruits.extend(cars)
print(fruits)
```

index(): Returns the index of the first element with the specified value
index()메소는 제일 앞의 원소를 리턴한다.

```python
fruits = ['appel', 'banana', 'cherry']
x = fruits.index("cherry")
y = fruits.index('banana')
#z = fruits.index('peach')   #안에 없으면 에러가 난다.
print(x)
print(y)
#print(z)
```

insert(): Adds an element at the specified position


```python
fruits = ['apple', 'banana', 'cherry']
fruits.insert(1, 'orange')
print(fruits)
```

pop(): Removes the element at the specified position
인덱스를 넣으면 그곳의 원소삭제 
디폴트는 -1 제일뒤 원소 삭제
그리고 삭제한 원소 리턴해준다.

```python
fruits = ['apple', 'banana', 'cherry']
fruits.pop(1)
print(fruits)
fruits.pop()
print(fruits)
```

remove(): Removes the item with the specified value
원소를 넣으면 리스트의 그 원소와 같은 가장 첫번째원소삭제

```python
fruits = ['apple', 'banana', 'cherry', 'banana']
fruits.remove("banana")
print(fruits)
```

reverse(): Reverses the order of the list
거꾸로 정렬해줌



```python
fruits = ['apple', 'banana', 'cherry']
fruits.reverse()     ## return 없음
x = fruits.reversed()## return 있음
print(fruits)
```

sort(): sort the list
정렬해준다.
123, abc순서대로

---
sort(reverse=플래그, key=정렬할 키값)



```python
cars = ['Ford', 'BMW', 'Volvo']
cars.sort()
print(cars)
```

```python
cars.sort(reverse=True)
print(cars)
```

```python
def myfunc(e):    #단어의 길이로 정렬
  return len(e)
cars = ['Ford', 'Mitsubishi', 'BMW', 'VW']
cars.sort(key=myfunc)
print(cars)
```

```python
def myFunc(e):
  return e['year']
cars = [
  {'car': 'Ford', 'year': 2005},
  {'car': 'Mitsubishi', 'year': 2000},
  {'car': 'BMW', 'year': 2019},
  {'car': 'VW', 'year': 2011}
]

cars.sort(key=myFunc)
print(cars)
```

count(): Returns the number of times a specified value occurs in a tuple
튜플에서 입력한 원소가 몇개나 중복되어있는지 확인

```python
thistuple = (1, 3, 7, 8, 7, 5, 4, 6, 8, 5)
x = thistuple.count(5)
print(x)
```

###실습. fruit_dict에서 과일명 찾기)
 과일 가격이 3000원 보다 큰 과일과 3000원 보다 작은 과일은?
1. high에 3000원보다 큰 과일이름
2. low에 3000원보다 작은 과일이름
3. 작업단계
    - high, low 리스트 빈방 만들기
    - for문으로 fruit_dict 내용을 하나씩 읽는다.
    - if문으로 과일가격 체크 3000원보다 큰것과 작은것
    + 크면 -> high에 추가
    + 작으면 -> low에 추가


```python
high = []
low = []
for i in fruit_dict.items():
  if(i[1] > 3000): high.append(i[0])
  else: low.append(i[0])
print("high = ", high)
print("low = ", low)
```

##추가 퀴즈

```python
#과일명 : [과일가격, 갯수]
fruit_count = {
    'apple' : [3500,10],
    'peer' : [2500,5],
    'cherry' : [5000,2]
}
```

3000원 보다 비싼 과일의 수는?

```python
high = []
row = []
row_num = 0
high_num = 0
for i in fruit_count.items():
  if(i[1][0] > 3000): 
    high.append(i[0])
    high_num += i[1][1]
  else:
    row.append(i[0])
    row_num+=i[1][1]
print("high = ", high)
print("row = ", row)
print("비싼 과일의 총 개수 = ", high_num)
print("싼 과일의 총 개수 = ", row_num)
```

```python
class_a = { "A1" : [78,60,98,88], "A2" : [66, 88, 77, 90], "A3" : [77, 93, 83, 73], "A4" : [55, 99, 86, 87] }
```

실습A 반 총 인원수?

```python
#A반은 총 몇명?
print(len(class_a))
```

실습2 학생별 총점과 평균은?

```python
total = []
avg = []
for i in class_a.items(): 
  total.append(sum(i[1]))
  avg.append(sum(i[1])/len(class_a))
  
for i in range(len(class_a)):
  print("A",i,"의 총점 = ",total[i], ", 평균 = ", avg[i])
```

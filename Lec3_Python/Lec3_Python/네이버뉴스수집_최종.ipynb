```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def get_page(url):
  header = {"User-Agent":"Mozilla.5.0"}
  html = requests.get(url, headers = header).text
  soup = BeautifulSoup(html, "html.parser")
  type06 = soup.find("ul", {"class" : "type06_headline"})
  return type06.find_all("dl")  #list 형으로 가져옴

def page_to_items(dl, day, page):
  tot_list = []
  for item2 in dl:
    #이미지 소스 가져오기
    try:  
      img = item2.find("dt", {"class" : "photo"}).find("img")
      img_url = img["src"]
    except:
      img_url = np.nan
    #뉴스 주소 가져오기
    try:
      img = item2.find("dt", {"class" : ""}).find("a")
      link_content_url = img["href"]
    except:
      link_content_url = np.nan
    #뉴스 타이틀 가져오기
    try:
      img = item2.find("dt", {"class" : ""}).text.replace("\t", "").replace("\n", "").replace("\r", "")
      link_title = img
    except:
      link_title = np.nan
    #뉴스 기사 요약 가져오기
    try:
      img = item2.find("span", {"class" : "lede"}).text
      content = img
    except:
      content = np.nan
    tot_list.append([pd.to_datetime(str(day), format='%Y%m%d'), page, img_url, link_content_url, link_title, content])  #날자는 타임스탬스
  return tot_list


import time #너무 빨리읽으면 공격으로 인식해서 딜레이 준다.
#time.sleep(0.5) #활용하기

#url 정하기
#읽어올 날짜, 시간
day = 20210609
page = 1

#기간
day_turm = 1
page_turm = 2

data_list = []
for i in range(0, day_turm+1):    #시작날부터 dayturm(1)후까지
  for j in range(0, page_turm):  #시작페이지부터 page_turm(2)까지
    url = "https://news.naver.com/main/list.nhn?mode=LSD&mid=sec&sid1=001&date=" + str(day+i) + "&page=" + str(page+j)  

    dl = get_page(url)  #dl가져오기
    print(day+i, page+j)
    data_list += page_to_items(dl, day+i, page+j) #읽어온 정보 리스트로 계속 합쳐줌
    time.sleep(0.5)


df = pd.DataFrame(data=data_list, columns=["날짜","페이지", "이미지 주소", "링크 주소", "타이틀", "기사내용"])
df.to_csv("/content/drive/MyDrive/Colab Notebooks/2021_06_10/naver_news2", index=False)
pd.read_csv("/content/drive/MyDrive/Colab Notebooks/2021_06_10/naver_news2")

#이미지 저장에 대해서는 2021_06_10/2021_06_10_오전2_데이터수집_함수_여러페이지_뉴스_수집하기 파일참조
```
